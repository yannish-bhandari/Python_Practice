# Exercise 1

Load the `iris.csv` dataset

Follow this instructions:
 1. Get all `Iris-setosa` flowers
 2. Get all flowers whose `sepal_length` is greater than 5cm and the `sepal_width` is lower than 2.5
 3. Create a new column with the ratio between petal and sepal length
 4. Count the number of flowers from each class with the ratio calculated in previous question greater than 0.5

```python
data = pd.read_csv("data/iris.csv")
data.head()
```

#### 1. Get all Iris-setosa flowers

```python
data[data["class"]=="Iris-setosa"]
```

#### 2. Get all flowers whose sepal_length is greater than 5cm and the sepal_width is lower than 4

```python
data[(data["sepal_length"]>5) & (data["sepal_width"]<2.5)]
```

#### 3. Create a new column with the ration between petal and sepal length

```python
data["ratio"] = data["petal_length"]/data["sepal_length"]
```

#### 4 Count the number of flowers from each class with the ratio calculated in previous question greater than 0.5

```python
print(f'Versicolor: {len(data.loc[(data["ratio"]>0.5) & (data["class"]=="Iris-versicolor")])}')
print(f'Virginica: {len(data.loc[(data["ratio"]>0.5) & (data["class"]=="Iris-virginica")])}')
print(f'Setosa: {len(data.loc[(data["ratio"]>0.5) & (data["class"]=="Iris-setosa")])}')
```
